```python
import pandas as pd
import numpy as np
```

```python
# IMPORTAR DATASET: TEMPERATURA DE SEATLE

features = pd.read_excel("/content/temps.xlsx")

features.head()
```

```python
features.describe()
```

```python
features.info()
```

```python
# TRANSFORMAR VARIÁVEL OBJECT:

features = pd.get_dummies(features)

features.head()
```

```python
# TRANSFORMAR VARIÁVEIS
# chamei nossa variável target de labels
# chamei a variável explicativa de todas colunas do df features, digo que quero todas as colunas do df, menos a coluna actual(minha label = y = target), também digo que quero excluir minhas colunas e não as linhas.


labels = np.array(features['actual'])

features = features.drop("actual", axis = 1)

feature_list = list(features.columns)

features = np.array(features)
```

```python
from sklearn.model_selection import train_test_split

train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.25, random_state=42)
# separei meus dados em 75% para treino de 25% para teste.
```

```python
baseline_preds = test_features[:, feature_list.index("average")]

baseline_error = abs(baseline_preds - test_labels)

print("Baseline error average:", round(np.mean(baseline_error), 2))
```

```python
from sklearn.datasets import make_moons
from sklearn import metrics
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
```

```python
# TREINAR O MODELO RandomForestRegresso

rf = RandomForestRegressor(n_estimators=1000, random_state=42)
rf.fit(train_features, train_labels)
```

```python
prediction_rf = rf.predict(test_features)

error_rf = abs(prediction_rf - test_labels)

r_sq = rf.score(features, labels)

print("R²:", r_sq)
print("MAE:", metrics.mean_absolute_error(test_labels, prediction_rf))
print("MSE:", metrics.mean_squared_error(test_labels, prediction_rf))
```

```python
# TREINAR O MODELO AdaBoostRegressor

ada = AdaBoostRegressor(n_estimators=100)
ada.fit(train_features, train_labels)
```

```python
ada_pred = ada.predict(test_features)

error_ada = abs(ada_pred - test_labels)

r_sq = ada.score(features, labels)

print("R²:", r_sq)
print("MAE:", metrics.mean_absolute_error(test_labels, ada_pred))
print("MSE:", metrics.mean_squared_error(test_labels, ada_pred))
```

```python
# TREINAR O MODELO GradientBoostingRegressor

gbr = GradientBoostingRegressor(n_estimators=100)
gbr.fit(train_features, train_labels)
```

```python
gbr_pred = gbr.predict(test_features)

error_gbr = abs(gbr_pred - test_labels)

r_sq = gbr.score(features, labels)

print("R²:", r_sq)
print("MAE:", metrics.mean_absolute_error(test_labels, gbr_pred))
print("MSE:", metrics.mean_squared_error(test_labels, gbr_pred))
```

```python
# REPRESENTAÇÃO DA ARVORE DE DECISÃO:
# 3 camadas de nós (max_depth=3)

rf = RandomForestRegressor(max_depth=3)
rf.fit(train_features, train_labels)

tree = rf.estimators_[5]

tree

from sklearn.tree import export_graphviz
import pydot
export_graphviz(tree, out_file= "tree.dot", feature_names = feature_list, rounded=True, precision = 1)

(graph, ) = pydot.graph_from_dot_file("tree.dot")
graph.write_png("tree.png")
```

```python
# QUAIS VARIÁVEIS SÃO MAIS RELEVANTES DENTRO DA MINHA ARVORE:
# FEATURE_IMPORTANCE: para avaliar quais variáveis são usadas na minha árvore (mais relevantes) e quais não são.
# foi colocar em ordem de relevância, nível de importÂncia. Nesse caso a variávei mais representativa para predizer a minha target, é a temperatura de um dia anterior.

importances = list(rf.feature_importances_)

feature_importance = [(feature, round(importance, 2)) for feature, importance in zip(feature_list, importances)]

feature_importance = sorted(feature_importance, key = lambda x: x[1], reverse=True)

[print("Feature: {:20} Importance {}".format(*pair)) for pair in feature_importance];
```
